```python
zip_path = '/home/wells/Documentos/Data_engineer/sp_transit_api_dataset.zip'
extract_dir = 'descompactado'
bucket_name = 'traffic-sp'
destination_prefix = 'bronze/'
```

```python
s3 = boto3.client('s3')
```

```python
print(f"descompactado {zip_path}...")
```

```python
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("Arquivos extraidos em:", extract_dir)
```

```python
for root, dirs, files in os.walk(extract_dir):
    for file in files:
        local_file_path = os.path.join(root, file)
        
        s3_key = destination_prefix + file
        
        print (f"Enviado{local_file_path} para s3://{bucket_name}/{s3_key}")
        
        s3.upload_file(local_file_path, bucket_name, s3_key)
        
    print("Upload concluido!")
```
